```python
import pandas as pd

# Read the three files into python as dataframes.

data_1 = pd.read_csv ("/Users/noachmeged/Documents/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv") 
data_2 = pd.read_csv ("/Users/noachmeged/Documents/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv")
data_3 = pd.read_csv ("/Users/noachmeged/Documents/Ironhack/Labs/lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv")
```

```python
# Show the DataFrame's shape.
data_1.shape

data_2.shape

data_3.shape
```

```python
# Standardize the header names.
data_1.columns = data_1.columns.str.lower().str.replace(" ", "_")
data_2.columns = data_2.columns.str.lower().str.replace(" ", "_")
data_3.columns = data_3.columns.str.lower().str.replace(" ", "_")
```

```python
# Rearrange the columns in the dataframe as needed
data_1.rename(columns={'st': 'state'}, inplace=True)
data_2.rename(columns={'st': 'state'}, inplace=True)
```

```python
# Concatenate the three dataframes
data_concat = pd.concat([data_1,data_2,data_3]).reset_index(drop=True)
data_concat
```

```python
# Which columns are numerical?
data_concat.dtypes 
# Based on the type and looking with the add functiom. I can tell that 
# number_of_open_complaints, customer_lifetime_value, income, monthy_premium_auto, total_claim_amount is numerical
```

```python
#Which columns are categorical?
# gender, education,policy_type, vehicle_class
```

```python
# Understand the meaning of all columns
#customer                      costumer id
#state                         state of the costumer
#gender                        sex of the costumer
#education                     level of education
#customer_lifetime_value       Customer lifetime value is the total worth to a business of a customer over the whole period of their relationship with the brand.
#income                        Yearly income of costumers in dollars
#monthly_premium_auto          The premium is the amount of money paid by the policyholder to the insurance company in exchange for coverage.
#number_of_open_complaints     The number of complaints by the costumer
#policy_type                   If the costumer owns a corporate auto or personal auto
#vehicle_class                 If the costumers owns an personal or corporate auto
#total_claim_amount            The total amount of money the costumer claimed by the insurance
```

```python
# Perform the data cleaning operations mentioned so far in class
#Delete the column education and the number of open complaints from the dataframe.
data_concat.drop(['education', 'number_of_open_complaints'], axis=1, inplace=True)
```

```python
data_concat
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 
data_concat['customer_lifetime_value'] = data_concat['customer_lifetime_value'] * 100

data_concat
```

```python
# Change dtype to numerical type.
data_concat['customer_lifetime_value'] = pd.to_numeric(data_concat['customer_lifetime_value'], errors='coerce')

data_concat.dtypes
```

```python
# Check for duplicate rows in the data and remove if any.
data_concat = data_concat.drop_duplicates()
data_concat
```

```python
# Filter out the data for customers who have an income of 0 or less.
data_concat = data_concat[data_concat['income'] <= 0]
data_concat
```
